```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('C:\\python-project\\data_files\\sample_data.xml')
root = tree.getroot() # Get the root element, which is <store> in our case

store_name = root.find('name').text
store_location = root.find('location').text

contact = root.find('contact')
phone = contact.find('phone').text
email = contact.find('email').text

product_data=[]
for product in root.find('products').findall('product'):
    product_info= {
        'id': product.find('id').text,
        'name': product.find('name').text,
        'category': product.find('category').text,
        'price': product.find('price').attrib.get('currency', 'USD'),
        'inStock': product.find('inStock').text,
    }

ratings = product.find('ratings')
if ratings is not None:
    product_info['average'] = ratings.find('average').text
    product_info['totalReviews'] = ratings.find('totalReviews').text

details = product.find('details')
if details is not None:
    product_info['color'] = details.find('color').text
    dimensions = product.find('dimensions')
    if dimensions is not None:
        product_info['length'] = dimensions.find('length').text
        product_info['width'] = dimensions.find('width').text
        product_info['height'] = dimensions.find('height').text

    product_info['store_name'] = store_name
    product_info['store_location'] = store_location
    product_info['phone'] = phone
    product_info['email'] = email


product_data.append(product_info)

for data in product_data:
    print(data)
```
