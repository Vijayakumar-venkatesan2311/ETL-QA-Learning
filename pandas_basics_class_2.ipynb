```python
# low_memory=True (default): 
#     When low_memory is set to True (or not specified), pandas reads the file in smaller chunks.This approach helps manage memory usage and 
# is useful for very large files. However, it can sometimes lead to inconsistent data types across chunks, which pandas tries to reconcile 
# after reading the entire file.
# low_memory=False: 
#     When low_memory is set to False, pandas reads the entire file into memory in one go. This ensures that data types are inferred once for the whole 
# dataset, leading to more consistent data types across the DataFrame. However, it requires enough RAM to hold the entire file, 
# which can be a limitation for very large files.
```
