```python
df5 = pd.DataFrame({'Name': ['Raj', 'Sethu', 'Rohit', 'Naresh']}, index=[1,2,3,4])
df6 = pd.DataFrame({ 'Score': [85, 90, 95, 89]}, index=[1,2,3,4])
```

```python
pd.merge(df5, df6, left_index=True, right_index=True)
```

```python
df7 = pd.DataFrame({'Name': ['Raj', 'Sethu', 'Rohit', 'Naresh']}, index=[1,2,3,4])
df8 = pd.DataFrame({ 'ID':[1,2,3,4], 'Score': [85, 90, 95, 89]})
```

```python
pd.merge(df7, df8, left_index=True, right_on='ID').set_index('ID')
```

```python
# If the DataFrames have multiple columns with the same name, pandas will throw an error unless you tell it which column to merge on explicitly 

# To overcome this we have suffix function

df9 = pd.DataFrame({'ID': [1, 2], 'Name': ['Alice', 'Bob'], 'Score': [85, 90]})
df10 = pd.DataFrame({'ID': [1, 2], 'Name': ['Alice', 'Bob'], 'Score': [88, 92]})

pd.merge(df9, df10, on='ID', suffixes=('_x', '_y'))
```
