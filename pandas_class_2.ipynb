```python
import pandas as pd
```

```python
# sorting data in a pandas dataframe - sort_values or sort_index
data = {
    'Name': ['David', 'Bob', 'Alice', 'Vikram', 'Charlie', 'Raj', 'Charlie'],
    'Age': [25, 45, 35, 50, 34, 29, 18],
    'Score': [88, 76, 95, 62, 70, 98, 90],
    'Passed': [True, True, True, False, True, True, True]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# sort by single column
df_sorted_single = df.sort_values(by='Name')
```

```python
df_sorted_single
```

```python
# sort by multiple column
df_sorted_multiple = df.sort_values(by=['Name', 'Score'])
```

```python
df_sorted_multiple
```

```python
df_sorted_2 = df.sort_values(by=['Name', 'Score'], ascending=[True, False])
```

```python
df_sorted_2
```

```python
df = pd.DataFrame(data, index=[2,4,6,1,3,5,0])
```

```python
df
```

```python
df.sort_index()
```

```python
df.sort_values(by=['Name', 'Score'], ascending=[True, False], inplace=True)
```

```python
df
```

```python
df.sort_index(ascending=False)
```
